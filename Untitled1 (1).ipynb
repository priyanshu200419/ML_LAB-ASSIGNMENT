```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, accuracy_score
import optuna
import matplotlib.pyplot as plt

# Load the wine dataset from a custom CSV file
df = pd.read_csv('/content/WineQT.csv')  # Replace '/content/WineQT.csv' with your file path

# Check if the dataset has a 'quality' column
if 'quality' not in df.columns:
    raise ValueError("The dataset must have a 'quality' column representing the target variable.")

# Features and target selection
X = df.drop(['Id', 'quality'], axis=1)
y = df['quality']

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define objective function for Optuna
def objective(trial):
    model_name = trial.suggest_categorical('model', ['RandomForest', 'Ridge', 'SVR'])

    if model_name == 'RandomForest':
        n_estimators = trial.suggest_int('n_estimators', 50, 200)
        max_depth = trial.suggest_int('max_depth', 3, 20)
        model = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth, random_state=42)

    elif model_name == 'Ridge':
        alpha = trial.suggest_loguniform('alpha', 1e-3, 10.0)
        model = Ridge(alpha=alpha)

    elif model_name == 'SVR':
        C = trial.suggest_loguniform('C', 1e-3, 10.0)
        epsilon = trial.suggest_loguniform('epsilon', 1e-3, 1.0)
        model = SVR(C=C, epsilon=epsilon)

    # Perform StratifiedKFold cross-validation
    n_splits = 5  # Number of splits for cross-validation
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

    # Initialize a list to store RMSE for each fold
    rmse_scores = []

    for train_index, test_index in skf.split(X_scaled, y):
        # Split data into training and testing sets for this fold
        X_train_fold, X_test_fold = X_scaled[train_index], X_scaled[test_index]
        y_train_fold, y_test_fold = y.iloc[train_index], y.iloc[test_index]

        # Train the model on the current fold
        model.fit(X_train_fold, y_train_fold)

        # Predict on the test set
        y_pred_fold = model.predict(X_test_fold)

        # Calculate RMSE for this fold
        rmse = mean_squared_error(y_test_fold, y_pred_fold, squared=False)
        rmse_scores.append(rmse)

    # Return the mean RMSE across all folds
    return np.mean(rmse_scores)

# Run Optuna optimization
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)

# Print the best trial
print("Best trial:")
print(f"  Value: {study.best_trial.value}")
print(f"  Params: {study.best_trial.params}")

# Extract the best model from the Optuna study
best_params = study.best_trial.params
model_name = best_params['model']

if model_name == 'RandomForest':
    best_model = RandomForestRegressor(
        n_estimators=best_params.get('n_estimators', 100),
        max_depth=best_params.get('max_depth', None),
        random_state=42
    )
elif model_name == 'Ridge':
    best_model = Ridge(alpha=best_params.get('alpha', 1.0))
elif model_name == 'SVR':
    best_model = SVR(
        C=best_params.get('C', 1.0),
        epsilon=best_params.get('epsilon', 0.1)
    )

# Train the best model on the entire dataset
best_model.fit(X_scaled, y)

# Predict on the test set
y_pred = best_model.predict(X_scaled)

# Calculate metrics (RMSE and accuracy)
rmse = np.sqrt(mean_squared_error(y, y_pred))
accuracy = accuracy_score(y, np.round(y_pred))

print(f"\nBest Model: {model_name}")
print(f"RMSE: {rmse:.2f}")
print(f"Accuracy: {accuracy:.2f}")

# Feature importance for RandomForest
if model_name == 'RandomForest':
    feature_importances = best_model.feature_importances_
    features = X.columns.values

    # Create a DataFrame for feature importance visualization
    importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})
    importance_df = importance_df.sort_values(by='Importance', ascending=False)

    # Plot the feature importance
    plt.figure(figsize=(10, 6))
    plt.bar(importance_df['Feature'], importance_df['Importance'], color='skyblue')
    plt.xticks(rotation=45, ha='right')
    plt.title('Feature Importance (Random Forest)', fontsize=16)
    plt.xlabel('Features', fontsize=14)
    plt.ylabel('Importance', fontsize=14)
    plt.tight_layout()
    plt.show()

# Scatter plot for Predicted vs Actual
plt.figure(figsize=(10, 6))

# Scatter plot of predicted vs actual
plt.scatter(y, y_pred, alpha=0.6, color='blue', label='Predicted Values')

# Add the perfect prediction line
plt.plot([min(y), max(y)], [min(y), max(y)], color='red', linestyle='--', linewidth=2, label='Perfect Prediction')

# Add title and labels
plt.title(f'Predicted vs Actual Quality Scores\nRMSE: {rmse:.2f}', fontsize=16)
plt.xlabel('Actual Quality', fontsize=14)
plt.ylabel('Predicted Quality', fontsize=14)

# Add grid and legend
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend(fontsize=12, loc='upper left')

# Display the plot
plt.tight_layout()
plt.show()
```
